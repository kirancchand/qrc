```python
from qiskit import QuantumCircuit, Aer, execute
import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt

# Parameters for the QRC
n_qubits = 4  # Number of qubits in the reservoir
n_steps = 100  # Number of time steps for training/testing
reservoir_depth = 10  # Depth of the reservoir (layers)

# Generate synthetic time series data
def generate_time_series(length, seed=42):
    np.random.seed(seed)
    return np.sin(np.linspace(0, 4 * np.pi, length)) + 0.1 * np.random.randn(length)

# Quantum Reservoir layer
def reservoir_layer(qc, params, x):
    for i in range(n_qubits):
        qc.ry(params[i] * x, i)
    for i in range(n_qubits - 1):
        qc.cz(i, i + 1)

# Create a quantum reservoir circuit
def create_quantum_reservoir(x, params):
    qc = QuantumCircuit(n_qubits)
    for layer_params in params:
        reservoir_layer(qc, layer_params, x)
    return qc

# Collect the reservoir states using statevector_simulator
def collect_reservoir_states_statevector(data, params):
    simulator = Aer.get_backend('statevector_simulator')
    states = []
    for x in data:
        qc = create_quantum_reservoir(x, params)
        qc.save_statevector()  # Save the statevector
        job = execute(qc, backend=simulator)
        result = job.result()
        statevector = result.get_statevector(qc)
        states.append(np.real(statevector[:2**n_qubits]))  # Take real part of the full statevector
    return np.array(states)

# Collect the reservoir states using qasm_simulator
def collect_reservoir_states_qasm(data, params):
    simulator = Aer.get_backend('qasm_simulator')
    states = []
    for x in data:
        qc = create_quantum_reservoir(x, params)
        qc.measure_all()  # Add measurement
        job = execute(qc, backend=simulator, shots=1024)
        result = job.result()
        counts = result.get_counts()
        # Normalize and take expectation values
        state = np.zeros(n_qubits)
        for bitstring, count in counts.items():
            for i, bit in enumerate(reversed(bitstring)):
                state[i] += int(bit) * count / 1024.0
        states.append(state)
    return np.array(states)

# Main execution
def main():
    # Generate data
    data = generate_time_series(n_steps)
    
    # Split into training and testing
    split = int(0.8 * n_steps)
    train_data, test_data = data[:split], data[split:]

    # Prepare inputs and targets
    X_train = train_data[:-1]
    y_train = train_data[1:]
    X_test = test_data[:-1]
    y_test = test_data[1:]

    # Generate random parameters for the reservoir
    params = np.random.uniform(0, 2 * np.pi, (reservoir_depth, n_qubits))

    # Collect reservoir states
    # Uncomment one of the lines below depending on the simulator you want to use:
    # train_states = collect_reservoir_states_statevector(X_train, params)
    # test_states = collect_reservoir_states_statevector(X_test, params)
    train_states = collect_reservoir_states_qasm(X_train, params)
    test_states = collect_reservoir_states_qasm(X_test, params)

    # Train a linear model (Ridge Regression)
    model = Ridge(alpha=1.0)
    model.fit(train_states, y_train)

    # Predict
    predictions = model.predict(test_states)

    # Plot results
    plt.figure(figsize=(10, 5))
    plt.plot(range(len(test_data)), test_data, label="True")
    plt.plot(range(1, len(test_data)), predictions, label="Predicted")
    plt.xlabel("Time step")
    plt.ylabel("Value")
    plt.legend()
    plt.title("Quantum Reservoir Computing - Time Series Prediction (Qiskit)")
    plt.show()

if __name__ == "__main__":
    main()

```
